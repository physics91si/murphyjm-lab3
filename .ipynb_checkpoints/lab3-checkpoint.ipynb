```python
# Edit this function. "pass" tells Python to do nothing.
def parse_sample(text_str, words_only=False, sort_list=False):
    parsed_text = text_str.split()
    for str in parsed_text:
        str = str.split('-')
    if words_only:
        for i in range(0, len(parsed_text)-1):
            parsed_text[i] = parsed_text[i].strip(":!?.-'\",;&()[]{}").lower()
    if sort_list:
        parsed_text.sort()
    return parsed_text
```
